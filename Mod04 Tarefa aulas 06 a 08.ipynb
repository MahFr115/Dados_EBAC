####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado ('Estado'ou Município), 
# e que devolve os dados da consulta no formato DataFrame.

def criar_df():
    #Consulta do url interesse 

    estado = input('Qual estado deseja consultar?\n').upper()
    interesse = input('Qual o interesse deseja consultar?\n').capitalize()
    
    # Gerando o URL 
    url_geral = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={estado}&tipo_interessado={interesse}'

    API = requests.get(url_geral)
    print('\n')
    print(API.status_code)
    print('\n')

    # Consulta dos dados
    data_json2 = API.json()

    # Criando o DataFrame do API
    PVL_DF = pd.DataFrame(data_json2['items'])
    
    return PVL_DF, url_geral
````

````python
# Testar o código para criar url

df_teste, url_teste = criar_df()

print(f'URL da consulta: {url_teste}')
````

````python
# 2) Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 
# 'Arquivado por decurso de prazo' estão registradas?

#Importar os dados de interesse
PVL_MG, url = criar_df()
PVL_MG
````

````python
# Contando o número de solicitações arquivadas 
PVL_MG[PVL_MG['status'] == 'Arquivado por decurso de prazo'].status.value_counts()
````

````python
# 3) Qual é o município da Bahia com mais solicitações deferidas?
# Consultar dados da Bahia 

PVL_BA_mun, url = criar_df()
PVL_BA_mun
````

````python
#Consultado de status
print(PVL_BA_mun['status'].unique())
````

````python
# TOP município da Bahia com mais solicitações deferidas

PVL_BA_mun[(PVL_BA_mun['status'] == 'Deferido (PVL-IF)') | (PVL_BA_mun['status'] == 'Deferido')].interessado.value_counts().nlargest(1, keep='all')
````

````python
# 4) Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

#Consultando os dados da Bahia com interesse Estado

PVL_BA_EST, url = criar_df()

# Salvando 
PVL_BA_EST.to_csv(f'df_{estado}_{interesse}.csv')
````
