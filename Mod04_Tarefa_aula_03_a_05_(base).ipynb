## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
#Para ficar mais fácil comparar os resultados seguintes, vou criar um DataFram para vizualização dos resultados
#Criando o Dataframe vazio
notas = pd.DataFrame()

#acrescentando cada coluna 
notas['Matemática'] = df_mat
notas['Portugês'] = df_por
notas['Geografia'] = df_geo 
notas['Inglês'] = df_ing 
notas['História'] = df_his
notas['Física'] = df_fis
notas['Química'] = df_qui

notas
````

````python
#Totais por prova
tt_mat = 24
tt_por = 18
tt_geo = 8
tt_ing = 8
tt_his = 8
tt_fis = 12
tt_qui = 12

tt_pro = pd.DataFrame([{'Matemática': tt_mat, 'Português': tt_por, 'Geografia': tt_geo,
                        'Inglês': tt_ing, 'História': tt_his, 'Física': tt_fis, 'Química': tt_qui}])

tt_pro ['Total'] = tt_pro.sum(axis = 1)

tt_pro
````

````python
# 1) Seu código aqui

#Calculando a médias dos alunos por prova

print('O percentual de acertos dos alunos em Matemática foi:', round(100*df_mat['Qt_acertos'].mean()/tt_mat, 2),'%')

print('O percentual de acertos dos alunos em Português foi:', round(100*df_por['Qt_acertos'].mean()/tt_por, 2),'%')

print('O percentual de acertos dos alunos em Geografia foi:', round(100*df_geo['Qt_acertos'].mean()/tt_geo, 2),'%')

print('O percentual de acertos dos alunos em Inglês foi:', round(100*df_ing['Qt_acertos'].mean()/tt_ing, 2),'%')

print('O percentual de acertos dos alunos em História foi:', round(100*df_his['Qt_acertos'].mean()/tt_his, 2),'%')

print('O percentual de acertos dos alunos em Fícisa foi:', round(100*df_fis['Qt_acertos'].mean()/tt_fis, 2),'%')

print('O percentual de acertos dos alunos em Química foi:', round(100*df_qui['Qt_acertos'].mean()/tt_qui, 2),'%')
````

````python
# 2) Seu código aqui

#Soma de cada DataFrame calculado por aluno

notas['Total'] = df_mat + df_por + df_geo + df_ing + df_his + df_fis +df_qui

notas
````

````python
# 3) Seu código aqui

#Calculando o porcentual geral por aluno: 
#Soma de todos os acertos (ac_tt) dividido pela quantidade total de pontos possíveis (90 pontos)

maxi = tt_pro.iloc[0, 7]

notas['perc_geral (%)'] = round(notas['Total']*100/maxi, 2)

notas
````

````python
# 4) Seu código aqui

# Nota de corte = 45. Quantos alunos tiveram nota maior que 45?

Nota_corte = 45

al_etapa2 = notas[notas['perc_geral (%)'] >= Nota_corte].shape[0]

print('Foram aprovados para a 2ª etspa do vestibular', al_etapa2, 'alunos.')
````

````python
#Base de dados utilizados do ACRE part 1
arquivo = 'https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/SIPNI/COVID/uf/uf%3DAC/part-00000-97829158-cc41-4c75-aec4-752344e6e938.c000.csv'
````

````python
#Lendo o arquivo e imprimindo a tabela de dados

vac_ac = pd.read_csv(arquivo, sep = ';')
vac_ac
````

````python
vac_ac.info()
````

````python
# 1) Sua solução aqui
#Quantas vacinas estão registradas nessa base

print('Foram registradas', vac_ac['document_id'].nunique(), 'aplicações.')
print('Foram registradas', vac_ac['vacina_codigo'].nunique(), 'vacinas diferentes.')
````

````python
# 2) Quantos pacientes foram vacinados?
#Contagem de paciente_id distintos

print('Foram vacinados',vac_ac['paciente_id'].nunique(), 'pacientes únicos.')
````

````python
# 3) Quantos pacientes únicos tomaram a primeira dose?
#Contagem de paciente distintos com a condicional da vacina ser apenas 1ª dose

pri_dose = vac_ac[vac_ac['vacina_descricao_dose'] == '1ª Dose']['paciente_id'].nunique()

print(pri_dose, 'pacientes únicos tomaram a primeira dose')
````

````python
# 3) solução alternativa
#Criando uma nova tabela com a exclusão dos dados de paciente_id repetidos

pri_dose_alt = vac_ac[vac_ac['vacina_descricao_dose'] == '1ª Dose'].drop_duplicates(
    subset='paciente_id')['paciente_id'].shape[0]

print(pri_dose_alt, 'pacientes únicos tomaram a primeira dose')
````

````python
# 4)Quantos pacientes com menos de 18 anos foram vacinados?
#Contagem de paciente distintos com a condicional de ter menos de 18 anos

pac_18 = vac_ac[vac_ac['paciente_idade'] < 18]['paciente_id'].nunique()

print(pac_18, 'pacientes vacinados tinham mais de 18 anos')
````

````python
# 5)Quantos estabelecimentos aplicaram vacina no Acre?
#Contagem de valores unicos de nome Fantasia na tabela

estab_uni = vac_ac['estalecimento_noFantasia'].nunique()

print(estab_uni, 'estabelecimentos aplicaram as vacinas no Acre.')
````

````python
# 4) a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma 
# lista com esses valores.

col_estab = ['estabelecimento_valor', 'estabelecimento_razaoSocial', 'estalecimento_noFantasia',
'estabelecimento_municipio_codigo', 'estabelecimento_municipio_nome', 'estabelecimento_uf']

print('As colunas que se referem ao estabelecimento são:')
col_estab
````
